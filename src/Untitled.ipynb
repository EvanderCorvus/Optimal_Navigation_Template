```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 1. Choose a range for x and y
x = np.arange(-10, 10, 0.1)
y = np.arange(-10, 10, 0.1)

# 2. Create a meshgrid
X, Y = np.meshgrid(x, y)

# 3. Calculate arctan2
Z = np.arctan2(-X, Y)

# 4. Plot the results (surface plot)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('arctan2(y,x)')
ax.set_title('Plot of arctan2(x,y)')
plt.show()

# 5. Also, plot a contour plot
fig = plt.figure()
ax = fig.add_subplot(111)
contour = ax.contour(X, Y, Z, 20, cmap='viridis')
ax.clabel(contour, inline=True, fontsize=10)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Contour Plot of arctan2(x,y)')
plt.show()
```
